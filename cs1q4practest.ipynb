```python
print("Simple Number Evaluation Program")

number_input = int(input("How many numbers do you want to enter? "))

# this asks the user how many numbers to enter

i, numbers_list = 0, []

# i is a variable, and numbers_list is the list that would store the inputted numbers

while i < number_input:
    # this will convert the inputs into a float to be used when solving the average
    numbers_list.append(float(input(f"Enter number {i + 1}: ")))
    # this asks the user to enter the number
    i += 1
    #this makes the program ask questions based on how many numbers the user wants to enter

if numbers_list:
    total_sum = sum(numbers_list) # this solves the sum of all input numbers
    print(f"The sum of the entered numbers is: {total_sum}")
    average = total_sum / len(numbers_list)
    # this solves the average of all input numbers
    print(f"The average of the entered numbers is: {average}")


    print("These are the results on whether you numbers are odd or even.")

    for num in numbers_list:
        if num % 2 == 0:
            print(f"Number {int(num)} is even.")

            #this shows that the number being referred to is even
        else:
            print(f"Number {int(num)} is odd.")

            #this shows that the number being referred to is odd
```
